```python
#importing the libararies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
# Connecting to the database Pass the connection string to a variable, conn_url
conn_url = 'postgresql://postgres:123@localhost/SQL5310_FinalProject'

# Create an engine that connects to PostgreSQL server
engine = create_engine(conn_url)

# Establish a connection
connection = engine.connect()
```

```python
# Table 1: State
state ="create table state(\
                state_id int PRIMARY KEY NOT NULL,\
                state_code varchar(2));"

connection.execute(state)
```

```python
# Table 2: City
city ="create table city(\
                city_id int PRIMARY KEY NOT NULL,\
                city_name varchar,\
                state_id int,\
                FOREIGN KEY (state_id) REFERENCES state(state_id));"

connection.execute(city)
```

```python
# Table 3: Geolocation
geolocation ="create table geolocation(\
                zip_id int PRIMARY KEY NOT NULL,\
                zipcode varchar,\
                city_id int,\
                state_id int,\
                FOREIGN KEY (state_id) REFERENCES state(state_id),\
                FOREIGN KEY (city_id) REFERENCES city(city_id));"

connection.execute(geolocation)
```

```python
# Table 4: Customer
customer ="create table customer(\
                customer_id varchar PRIMARY KEY NOT NULL,\
                zip_id int,\
                FOREIGN KEY (zip_id) REFERENCES geolocation(zip_id));"

connection.execute(customer)
```

```python
# Table 5: Sellers
sellers ="create table sellers(\
                seller_id varchar PRIMARY KEY NOT NULL,\
                zip_id int,\
                FOREIGN KEY (zip_id) REFERENCES geolocation(zip_id));"

connection.execute(sellers)
```

```python
# Table 6: Product Category
product_category ="create table product_category(\
                product_category_id int PRIMARY KEY NOT NULL,\
                product_category_name varchar,\
                product_category_name_english varchar);"

connection.execute(product_category)
```

```python
# Table 7: Products
products ="create table products(\
                product_id varchar PRIMARY KEY NOT NULL,\
                product_category_id int,\
                product_description_length int,\
                product_name_length int,\
                product_photos_qty int,\
                product_weight_g numeric,\
                product_length_cm numeric,\
                product_height_cm numeric,\
                product_width_cm numeric,\
                FOREIGN KEY (product_category_id) REFERENCES product_category(product_category_id));"

connection.execute(products)
```

```python
# Table 8: Order list
order_list ="create table order_list(\
                order_id varchar PRIMARY KEY NOT NULL,\
                customer_id varchar,\
                FOREIGN KEY (customer_id) REFERENCES customer(customer_id));"

connection.execute(order_list)
```

```python
# Table 9: Order details
order_details ="create table order_details(\
                order_id varchar ,\
                order_item_id int,\
                product_id varchar,\
                seller_id varchar,\
                price numeric,\
                freight_value numeric,\
                PRIMARY KEY (order_id , order_item_id),\
                FOREIGN KEY (order_id ) REFERENCES order_list(order_id),\
                FOREIGN KEY (seller_id ) REFERENCES sellers(seller_id),\
                FOREIGN KEY (product_id ) REFERENCES products(product_id));"

connection.execute(order_details)
```

```python
# Table 10: Payment type
payment_type ="create table payment_type(\
                payment_type_id int PRIMARY KEY NOT NULL,\
                payment_type varchar);"

connection.execute(payment_type)
```

```python
# Table 11: Payment details
payment_details ="create table payment_details(\
                payment_id int PRIMARY KEY NOT NULL,\
                order_id varchar,\
                payment_type_id int,\
                payment_sequential int,\
                payment_installments int,\
                payment_value numeric,\
                FOREIGN KEY (order_id) REFERENCES order_list(order_id),\
                FOREIGN KEY (payment_type_id) REFERENCES payment_type(payment_type_id));"

connection.execute(payment_details)
```

```python
# Table 12: Delivery
delivery ="create table delivery(\
                order_id varchar PRIMARY KEY NOT NULL,\
                order_approved_at timestamp,\
                order_purchase_timestamp timestamp,\
                order_delivery_carrier_date timestamp,\
                shipping_limit_date timestamp,\
                order_delivery_customer_date timestamp,\
                order_estimated_delivery_date timestamp,\
                FOREIGN KEY (order_id) REFERENCES order_list(order_id));"

connection.execute(delivery)
```

```python
# Table 13: Order Status
order_status ="create table order_status(\
                order_id varchar PRIMARY KEY NOT NULL,\
                customer_id varchar,\
                payment_id int,\
                order_status varchar,\
                FOREIGN KEY (order_id) REFERENCES order_list(order_id),\
                FOREIGN KEY (customer_id) REFERENCES customer(customer_id),\
                FOREIGN KEY (payment_id) REFERENCES payment_details(payment_id));"

connection.execute(order_status)
```

```python
# Table 14: Review Details
review_details ="create table review_details(\
                review_id varchar,\
                order_id varchar,\
                review_score int,\
                review_comment_title varchar,\
                review_comment_message varchar,\
                review_creation_date timestamp,\
                review_answer_timestamp timestamp,\
                PRIMARY KEY (review_id , order_id),\
                FOREIGN KEY (order_id) REFERENCES order_list(order_id));"

connection.execute(review_details)
```

```python
#query 1: Does delivery time span (eg. A few days vs. a few weeks) affect the number of items ordered or total income per order?

queryCmd1 = """SELECT 
                    order_id, 
                    (order_delivery_customer_date - order_approved_at) AS delivery_time, 
                    MAX(order_item_id) AS order_quantity, SUM(price) AS total_order_value
                FROM delivery 
                NATURAL JOIN order_details WHERE delivery.order_id = order_details.order_id
                GROUP BY delivery.order_id
                ORDER BY order_quantity DESC, total_order_value DESC;"""

##natural join to only analyze complete records; use MAX not SUM/COUNT in order_item_id based on data structure/how company inputted data and identifying quantity of products ordered (may be same or different product)


#execute statement and get the results
results1 = connection.execute(queryCmd1).fetchall()

#extract column names
column_names1 = results1[0].keys()

#store results in a Dataframe
df_query1 = pd.DataFrame(results1, columns = column_names1)
```

```python
#query 2: How to target advertisement in each area (zipcode/state) to maximize marketing efficiency/minimize marketing cost?
queryCmd2 = """SELECT state_code, SUM(order_quantity) AS total_order_quantity_per_state, SUM(total_order_value) AS total_order_value_per_state
FROM (
        SELECT customer.customer_id, order_details.order_id, products.product_id, product_category.product_category_name, customer.zip_id, geolocation.state_id, state.state_code, product_category.product_category_name_english, COUNT(order_details.order_item_id) AS order_quantity, SUM(order_details.price) AS total_order_value
        FROM customer 
        JOIN order_list ON customer.customer_id = order_list.customer_id
        JOIN order_details ON order_list.order_id = order_details.order_id
        JOIN products ON order_details.product_id = products.product_id
        JOIN product_category ON products.product_category_id = product_category.product_category_id
        JOIN geolocation ON customer.zip_id = geolocation.zip_id
        JOIN state ON geolocation.state_id = state.state_id
        GROUP BY order_details.order_id, customer.customer_id, products.product_id, product_category.product_category_name, customer.zip_id, geolocation.state_id, state.state_code, product_category.product_category_name_english
        ORDER BY order_id DESC, product_id DESC) AS FOO
GROUP BY state_code
ORDER BY total_order_value_per_state DESC, total_order_quantity_per_state DESC;"""

#execute statement and get the results
results2 = connection.execute(queryCmd2).fetchall()

#extract column names
column_names2 = results2[0].keys()

#store results in a Dataframe
df_query2 = pd.DataFrame(results2, columns = column_names2)
```

```python
#query 6: which categories of products do customers buy together?
queryCmd6 =  """CREATE VIEW order_product_categories AS 
        SELECT DISTINCT FOO1.order_id, FOO1.product_id, product_category.product_category_name,  product_category.product_category_name_english
        FROM (
            SELECT DISTINCT order_id, product_id
            FROM order_details
            GROUP BY order_id, product_id) AS FOO1
        JOIN products ON FOO1.product_id = products.product_id
        JOIN product_category ON products.product_category_id = product_category.product_category_id; 

    SELECT COUNT (order_id), COUNT (product_category_name_english), CASE WHEN COUNT(order_id) = COUNT (product_category_name_english) THEN 'No client bought different product categories in one order' ELSE 'Client bought different product categories in an order' END AS multiple_products_per_order 
    FROM (
            SELECT DISTINCT order_id, product_category_name_english
            FROM order_product_categories
            GROUP BY order_id, product_category_name_english 
            ORDER BY order_id DESC) AS FOO2;

    SELECT * FROM order_product_categories 
    WHERE order_id = 
    (SELECT order_id 
    FROM (
        SELECT order_id, COUNT(DISTINCT order_id) AS product_category_count_per_order, product_category_name_english 
        FROM order_product_categories
        GROUP BY order_id, product_category_name_english 
        HAVING COUNT(DISTINCT order_id) >1) AS FOO3);
"""

#execute statement and get the results
results6 = connection.execute(queryCmd6).fetchall()

#extract column names
column_names6 = results6[0].keys()

#store results in a Dataframe
df_query6 = pd.DataFrame(results6, columns = column_names6)
```

```python
#query 8: what is the average freight cost for each city?
queryCmd8 = """select city_name, avg(freight_value) as avg_freight
from order_details
left join products
on products.product_id =order_details.product_id
left JOIN order_list
ON order_details.order_id=order_list.order_id
left JOIN customer
ON customer.customer_id=order_list.customer_id
left JOIN geolocation
ON geolocation.zip_id=customer.zip_id
left JOIN city
ON geolocation.city_id=city.city_id
group by city_name
order by avg_freight DESC;"""

#execute statement and get the results
results8 = connection.execute(queryCmd8).fetchall()

#extract column names
column_names8 = results8[0].keys()

#store results in a Dataframe
df_query8 = pd.DataFrame(results8, columns = column_names8)
```

```python
#query 9 Can our client reduce delivery cost? 
queryCmd9 ="""CREATE VIEW freight_customer_seller AS 
        SELECT order_details.order_id, customer.customer_id, sellers.seller_id, customer.zip_id AS customer_zip_id, sellers.zip_id AS sellers_zip_id, order_details.freight_value
        FROM order_details
        JOIN order_list ON order_details.order_id = order_list.order_id 
        JOIN sellers ON order_details.seller_id = sellers.seller_id
        JOIN customer ON order_list.customer_id = customer.customer_id; 

CREATE VIEW customer_freight AS 
        SELECT customer_zip_id, city.city_name AS customer_city_name, sellers_zip_id, freight_value, geolocation.city_id
        FROM freight_customer_seller
        JOIN geolocation ON customer_zip_id = geolocation.zip_id
        JOIN city ON geolocation.city_id = city.city_id;

CREATE VIEW sellers_freight AS 
        SELECT sellers_zip_id, city.city_name AS sellers_city_name
        FROM freight_customer_seller
        JOIN geolocation ON sellers_zip_id = geolocation.zip_id
        JOIN city ON geolocation.city_id = city.city_id;

SELECT customer_city_name, sellers_city_name, freight_value
FROM customer_freight
JOIN sellers_freight ON customer_freight.sellers_zip_id = sellers_freight.sellers_zip_id
ORDER BY freight_value DESC;
"""

#execute statement and get the results
results9 = connection.execute(queryCmd9).fetchall()

#extract column names
column_names9 = results9[0].keys()

#store results in a Dataframe
df_query9 = pd.DataFrame(results9, columns = column_names9)
```
